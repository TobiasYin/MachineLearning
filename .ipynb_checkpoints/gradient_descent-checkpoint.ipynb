```python
def dJ(theta):
    return 2*(theta - 2.5)
```

```python
def J(theta):
    return (theta - 2.5)**2 - 1
```

```python
theta = 0.0
eta = 0.6
theta_history = []
while True:
    gradient = dJ(theta)
#     if abs(gradient) - 0.0 <1e-8:
#         break
    theta_history.append(theta)
    last_theta = theta
    theta -=  eta * gradient
    if abs(J(theta) - J(last_theta)) < 1e-8:
        break
plt.plot(plt_x, plt_y)
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r')
plt.show()
theta
print(theta)
len(theta_history)
```

```python
theta = 0.0
eta = 0.1
theta_history = []
while True:
    gradient = dJ(theta)
    theta_history.append(theta)
    if abs(gradient) - 0.0 <1e-8:
        break
    theta -=  eta * gradient
    if len(theta_history)>3:
        break
plt.plot(plt_x, plt_y)
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r')
plt.show()
print(theta)
len(theta_history)
```

```python
def gradient_descent(initial_theta, eta, dJ = None, J = None, n_iters = 1e4, epsilon = 1e-8):
    theta = initial_theta
    theta_history = [theta]
    for i in range(int(n_iters)):
        gradient = dJ(theta)
        last_theta = theta
        theta -= eta * gradient
        theta_history.append(theta)
        if(abs(J(theta) - J(last_theta)) < epsilon):
            break
    return theta, theta_history
```

```python
def plot_theta_history(theta_history, J, method = "plot"):
    plt_x = np.linspace(min(theta_history[0] * 2,theta_history[0] * (-2)),max(theta_history[-1] * 2, theta_history[-1] * (-2)))
    plt_y = J(plt_x)
    plt.plot(plt_x, plt_y)
    if method == "plot":
        plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r')
    elif method == "scatter":
        plt.scatter(np.array(theta_history), J(np.array(theta_history)), color = 'r')
    plt.show()
```

```python
plot_theta_history(gradient_descent(0, 0.2, dJ, J)[1], J , method = "scatter")
```
