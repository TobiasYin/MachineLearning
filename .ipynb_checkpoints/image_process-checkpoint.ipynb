```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import imageio
import PIL
def tiny_image(filename:str, persent:float)->np.ndarray:
    im = PIL.Image.open(filename)
    im_array = [np.array(im)[:, :, i].T for i in range(3)]
    U, s, VT = [], [], []
    for dim in im_array:
        svd_result = np.linalg.svd(dim)
        U.append(svd_result[0])
        s.append(svd_result[1])
        VT.append(svd_result[2])
    S = []
    for i in s:
        S.append(np.zeros((len(U[1]),len(VT[1]))))
    percentile = np.percentile(s[0],persent)
    for i in range(len(s[1])):
        for j in range(len(S)):
            S[j][i][i] = s[j][i]
        if s[1][i] < percentile:
            break
    res = np.array([U[i].dot(S[i]).dot(VT[i]) for i in range(len(U))],dtype = "uint8")
    return res.T
def tiny_save(input_name:str, output_name:str, persent:float):
    res = tiny_image(input_name, persent)
    imageio.imwrite(output_name, res)
```

```python
tiny_save("test.jpg","0.jpg",0)
```
